# Problem 1

```python
import pandas as pd
import numpy as np
df = pd.read_html('http://www.collegesimply.com/colleges/california/',encoding='')[0]
df
```

# Problem 2

```python
num_of_col = df.groupby('Location').count()
num_of_col=num_of_col.reset_index()
```

# Problem 3

```python
num_of_col['City'] = num_of_col['Location'].apply(lambda x: ' '.join(x.strip().split()[:-1]).strip(','))
num_of_col
```

```python
url = "https://en.wikipedia.org/wiki/List_of_cities_and_towns_in_California"
ca_cities = pd.read_html(url, header=0, index_col=0)[1]
ca_cities.columns = ['Type','County','Population','sqmi','km2','Incorporated']
ca_cities = ca_cities.drop('sq mi')
ca_cities['Incorporated'] = ca_cities['Incorporated'].str.slice(23,)
ca_cities
```

```python
summ = pd.merge(ca_cities,num_of_col,left_index=True,right_on='City',how='left').drop('City',axis=1)
summ
```

# Problem 4
